```python

import pandas as pd
import json

# Укажите пути к файлам
input_json_file = 'input.json'    # Ваш входной JSON-файл
output_excel_file = 'output.xlsx' # Имя выходного Excel-файла

# Читаем JSON-файл
with open(input_json_file, 'r', encoding='utf-8') as f:
    data = json.load(f)
data_n = data.get("result")
data1 = data_n.get('operations')
# for a in data1:
#   print(a)
# headers = data_n[0].keys()
# rows = [list(d.values()) for d in data_n]
```

```python
# Создаём DataFrame
df = pd.DataFrame(data1)
```

```python
df['operation_type_name'].unique()
```

```python
df['operation_type_name'].value_counts()
```

```python
df['name'] = df['items'].apply(lambda x: x[0]['name'] if x and len(x) > 0 and 'name' in x[0] else None)
df['sku'] = df['items'].apply(lambda x: x[0]['sku'] if x and len(x) > 0 and 'sku' in x[0] else None)
df = df.drop(columns=['items'])
```

```python
df['schema'] = df['posting'].apply(lambda x: x['delivery_schema'] if isinstance(x, dict) and 'delivery_schema' in x else None)
df['order_date'] = df['posting'].apply(lambda x: x['order_date'] if isinstance(x, dict) and 'order_date' in x else None)
df['posting_number'] = df['posting'].apply(lambda x: x['posting_number'] if isinstance(x, dict)  and 'posting_number' in x else None)
df['warehouse_id'] =df['posting'].apply(lambda x: x['warehouse_id'] if isinstance(x, dict) and 'warehouse_id' in x else None)
df = df.drop(columns=['posting'])
```

```python
# Выводим таблицу
df.head(10)
```

```python
df.to_excel(output_excel_file, index=False)
print(f"DataFrame успешно сохранен в файл '{output_excel_file}'")
```
