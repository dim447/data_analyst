```python
# Импортируем файл с данными и смотрим первые 10 строк
import pandas as pd

df = pd.read_csv('/content/sample_data/winemag-data_first150k.csv')
df.head(10)
```

```python
# <перечень названий столбцов таблицы df>
df.columns
```

```python
# поменяем названия столбцов для наглядности и выведим случайные 10 строк
df.columns = ['Страна', 'Описание', 'Наименование', 'Оценка', 'Цена', 'Провинция',
       'Регион', 'Район', 'Сорт вина', 'Винодельня']
df.sample(10)
```

```python
# Посмотрим структуру Датафрейма
df.info()
```

```python
# Получим общее число строк
df.shape
```

```python
# посмотрим статистику (берется только по числовым данным)
df.describe()
```

```python
### Ищем нулевые ячейки. Видим, что не все данные заполнены,
# есть много пропусков в цене, регионе, наименовании и районе.
df.isnull().sum().sort_values()
```

```python
# можем делать разные отборы: по стране (Испания) и цене больше 150
df[(df['Страна'] == 'Spain') & (df['Цена'] > 150)]
```

```python
# Посмотрим самый популярный сорт винограда (количество провинций)
df.groupby('Сорт вина')['Провинция'].count().sort_values(ascending=False).head(20)
```

```python
# посмотрим какие винодельни выращивают 'Cabernet Franc' только в Калифорнии
df[(df['Сорт вина'] == 'Cabernet Franc') & (df['Провинция'] == 'California')]
```
