```python
### Загружаем библиотеку данных pandas и считываем данные из файла продаж
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sales_v2023-07-09.csv')
df.head()
```

```python
### Получаем данные о полях и их типе
df.info()
```

```python
### Получаем данные из таблицы с возвратами и смотрим на поля таблицы и типы
df_returns = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/returns_v2023-07-09.csv')
df_returns.info()
```

```python
### Смотри случайные 5 строк
df_returns.sample(5)
```

```python
### Делаем слияние двух таблиц по полю "order_id"
new_df = df.merge(df_returns, on='order_id')
```

```python
### Проверяем как прошло слияние: добавилось поле 'return'
new_df.head()
```

```python
new_df.groupby('region')['return'].sum().sort_values(ascending=False)
```

```python
### Проверяем на нулевые значения новую таблицу
new_df.isnull().sum().sort_values()
```

```python

# Посмотрим какая категория товара больше продается и отсортируем выручку по убыванию
df_category = new_df.groupby('category')['sales'].sum()
df_category.sort_values(ascending=False)
```

```python
### Сортируем таблицу по категории "Одежда и обувь"
df_shoe = new_df[new_df['category'] == 'Одежда и обувь']
df_shoe.head()
```

```python
df_shoe.info()
```

```python
# Посмотрим какай регион больше покупает "Одежда и обувь" и отсортируем выручку по убыванию
df_shoe.groupby('region')['sales'].sum().sort_values(ascending=False)
```

```python
# Посмотрим какай регион больше покупает (количество заказов) в категории "Одежда и обувь"
#и отсортируем количество по убыванию
df_shoe.groupby('region')['sales'].count().sort_values(ascending=False)
```

```python
# Вычисляем долю возвратов по регионам в категории "Одежда и обувь".
df_shoe_return = df_shoe.groupby('region')['return'].sum().sort_values(ascending=False)
result = []
for i in df_shoe_return:
  procent = round((i / df_shoe_return.sum()), 2)
  result.append(procent)


df_shoe_return = df_shoe_return.to_frame().reset_index()
df_shoe_return['fraction'] = result
df_shoe_return
```

```python
### Сортируем таблицу по категории "Бытовая техника"
df_technik = new_df[new_df['category'] == 'Бытовая техника']
df_technik.head()
```

```python
df_technik.info()
```

```python
# Посмотрим какай регион больше покупает по категории "Бытовая техника" и отсортируем выручку по убыванию
df_technik.groupby('region')['sales'].sum().sort_values(ascending=False)
```

```python
# Посмотрим какай регион больше покупает (количество заказов) в категории "Бытовая техника"
#и отсортируем количество по убыванию
df_technik.groupby('region')['sales'].count().sort_values(ascending=False)
```

```python
# Вычисляем долю возвратов по регионам в категории "Бытовая техника".
df_technik_return = df_technik.groupby('region')['return'].sum().sort_values(ascending=False)
result = []
for i in df_technik_return:
  procent = round((i / df_technik_return.sum()), 2)
  result.append(procent)


df_technik_return = df_technik_return.to_frame().reset_index()
df_technik_return['fraction'] = result
df_technik_return
```
