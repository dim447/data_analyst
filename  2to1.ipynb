# **Задачи исследования**

### Загружаем данные и меняем с 2 пачки на 1.




```python
### Загружаем данные продаж за период (orders.csv)
import pandas as pd
import numpy as np
from datetime import datetime


df = pd.read_csv('/content/orders.csv', sep=';')
df.info()
```

```python
### Убираем товары со статусом "Отменен"
df = df.loc[df['Статус'] != 'Отменён']
```

```python
### приводим значения артикулов с формату "Моя торговля"
df['Артикул'] = df['Артикул'].apply(lambda x: str(x)[:5] if len(str(x)) == 6
                                    else str(x)[:3] if len(str(x)) == 4
                                    else str(x))
```

```python
### Оставляем для импорта только эти данные
df_new = df[['Наименование товара', 'Артикул', 'Итоговая стоимость товара', 'Количество']]
```

```python
### Если в названии есть "2 пачки", то меняем количество на 2 и стоимость делим на 2
sub ='2 пачки'
df_new["Indexes"]= df_new['Наименование товара'].str.find(sub)
df_new.loc[df_new["Indexes"] >= 0, 'Количество'] = 2
df_new.loc[df_new["Indexes"] >= 0, 'Итоговая стоимость товара'] = df['Итоговая стоимость товара'] / 2
df_new.head(10)
```

```python
### Сохраняем в файл для импорта
df_new.to_excel('/content/orders_new.xlsx')
# df_new.to_csv('/content/orders_new.csv')
```
