### Загружаем данные и меняем с 2 пачки на 1.




```python
### Загружаем данные и объединяем семь датасетов
import pandas as pd
import numpy as np
from datetime import datetime


df = pd.read_csv('/content/orders.csv', sep=';')
```

```python
df.head(5)
```

```python
df_new = df[['Наименование товара', 'Артикул', 'Итоговая стоимость товара', 'Количество']]
df_new.head(5)
```

```python
sub ='2 пачки'
df_new["Indexes"]= df_new['Наименование товара'].str.find(sub)
df_new.loc[df_new["Indexes"] >= 0, 'Количество'] = 2
df_new
```

```python
df_new['art_str'] = df_new['Артикул'].astype(str)
# df_new['len_new_art'] = df_new['art_str'].str.len()
df_new["art_str"] = df_new["art_str"].str[:5]
df_new['Артикул'] = df_new['art_str'].astype('int32')
df_new
```
